```python

import numpy as np
import matplotlib.pyplot as plt

# Define the constants a and b
a = 2
b = 1

# Create a range of x values
x = np.linspace(-5, 5, 100)

# Calculate the corresponding y values using the hyperbola equation
y = np.sqrt(b**2 * (x**2/a**2 - 1))

# Plot the hyperbola
plt.plot(x, y, 'r', label='Hyperbola')
plt.plot(x, -y, 'r')  # Plot the other branch of the hyperbola

plt.xlabel('x')
plt.ylabel('y')
plt.title('Hyperbola: x^2/a^2 - y^2/b^2 = 1')
plt.legend()
plt.grid(True)
plt.show()
```
